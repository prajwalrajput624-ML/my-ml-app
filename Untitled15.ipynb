```python
# convert last updated column in datetime datatype
df['last_updated'] = pd.to_datetime(df['last_updated'],errors='coerce')
```

```python
# remove / in timezone column
df['timezone'] = df['timezone'].str.replace("/"," ",regex=False)
```

```python
# convert sunrise,sunset,moonrise,moonset columns convert in datetime dtypes
df['sunrise'] = pd.to_datetime(df['sunrise'],errors='coerce',format='mixed')
df['sunset'] = pd.to_datetime(df['sunset'],errors='coerce',format='mixed')
df['moonrise'] = pd.to_datetime(df['moonrise'],errors='coerce',format='mixed')
df['moonset'] = pd.to_datetime(df['moonset'],errors='coerce',format='mixed')
```

```python
# create a hour and mintue column using sunrise,sunset,moonrise,moonset columns
for col in ['sunrise','sunset']:
    df[col +"hour"] = df[col].dt.hour
    df[col +"minute"] = df[col].dt.minute
```

```python
df['day_length'] = (df['sunset'] - df['sunrise']).dt.total_seconds()/3600
```

```python
df.drop(columns=['sunset','sunrise','moonset','moonrise'],inplace=True)
```

```python
# convert all category column in category datatypes
cols = df.select_dtypes(include=['object']).columns
df[cols] = df[cols].apply(lambda x : x.astype('category'))
```

```python
# convert all category column in lowercase
category_columns = df.select_dtypes(include=['category']).columns
df[category_columns] = df[category_columns].apply(lambda x:x.str.lower()) 
```

### Model building

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix,classification_report,roc_auc_score
```

```python
def map_weather_class(condition):
    condition = condition.lower()
    if ('rain' in condition or
        'drizzle' in condition or
        'shower' in condition or
        'thunder' in condition or
        'thundery' in condition):
        return 'Rain'
    else:
        return 'Sunny'
```

```python
df['condition_grouped'] = df['condition_text'].apply(map_weather_class)
print(df['condition_grouped'].value_counts())
```

```python
df['condition_grouped'] = (df['condition_grouped'].str.strip().map({
        "Sunny" : 0,
        "Rain" : 1
    })
)
```

```python
categorical_columns = [
    "country",
    "location_name",
    "timezone",
    "wind_direction",
    "moon_phase"
]
numeric_columns = [
    "latitude",
    "longitude",
    "last_updated_epoch",
    "temperature_celsius",
    "temperature_fahrenheit",
    "wind_mph",
    "wind_kph",
    "wind_degree",
    "pressure_mb",
    "pressure_in",
    "precip_mm",
    "precip_in",
    "humidity",
    "cloud",
    "feels_like_celsius",
    "feels_like_fahrenheit",
    'visibility_km',
    'visibility_miles',
    'uv_index', 
    'gust_mph', 
    'gust_kph', 
    'air_quality_Carbon_Monoxide',
    'air_quality_Sulphur_dioxide', 'air_quality_PM2.5', 'air_quality_PM10',
       'air_quality_us-epa-index', 'air_quality_gb-defra-index',
    'moon_illumination'
]
```

```python
x = df[categorical_columns + numeric_columns]
y = df['condition_grouped']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
preprocessor = ColumnTransformer(transformers=[
    ("numeric_columns",StandardScaler(),numeric_columns),
    ("categorical_columns",OneHotEncoder(handle_unknown='ignore'),categorical_columns)
])
```

```python
pipeline = Pipeline(steps=[
    ("preprocessor",preprocessor),
    ("classifier",XGBClassifier(n_estimators=300, learning_rate=0.05, max_depth=6, subsample=0.8, colsample_bytree=0.8, gamma=1, min_child_weight=3,random_state=42, n_jobs=-1, eval_metric="logloss"))
])
```

```python
pipeline.fit(x_train,y_train)
```

```python
y_pred = pipeline.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
y_prob = pipeline.predict_proba(x_test)[:,1]
y_pred = (y_prob >= 0.35).astype(int)
```

```python
print(classification_report(y_test,y_pred))
```

```python
roc_auc = roc_auc_score(y_test,y_prob)
```

```python
print("Roc_Auc_Score Is --> ",roc_auc)
```

```python
# check training and testing score
print(pipeline.score(x_test,y_test))
print(pipeline.score(x_train,y_train))
```
