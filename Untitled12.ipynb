```python
# let's start the feature engineering
x = df.drop(columns=['time'],axis=1)
y = df['time']
```

```python
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score,KFold,GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
df.head()
```

```python
num_col = [
    "total_bill",
    'tip',
    'size'
]
cat_col = [
    "sex",
    'smoker',
    'day'
]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# define the ColumnTransformer
preprocess = ColumnTransformer([
    ('num_col',StandardScaler(),num_col),
    ('cat_col',OneHotEncoder(drop='first',handle_unknown='ignore'),cat_col)
])
```

```python
# define the pipeline
pipe = Pipeline([
    ('preprocess',preprocess),
    ('model',LogisticRegression())
])
```

```python
# fit the model
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
# Let's check the metrics
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix
```

```python
print("Accuracy score is -->",accuracy_score(y_test,y_pred))
```

```python
print("Precision score of dinner is -->",precision_score(y_test,y_pred,pos_label='dinner'))
```

```python
print("Precision score of lunch is -->",precision_score(y_test,y_pred,pos_label='lunch'))
```

```python
print("Recall score of dinner is -->",recall_score(y_test,y_pred,pos_label='dinner'))
```

```python
print("Recall score of lunch is -->",recall_score(y_test,y_pred,pos_label='lunch'))
```

```python
print("F1 score of dinner is -->",f1_score(y_test,y_pred,pos_label='dinner'))
```

```python
print("F1 score of lunch is -->",f1_score(y_test,y_pred,pos_label='lunch'))
```

```python
print("Confusion matrix -->",confusion_matrix(y_test,y_pred))
```

```python
# Let's check of cross val score
scores = cross_val_score(pipe,x,y,cv=kf)
```
