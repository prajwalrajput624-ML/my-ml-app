```python
num_cols = [
    'tenure',
    'TotalCharges',
    'MonthlyCharges'
]
cat_cols = [
    'gender',
    'SeniorCitizen',
    'Contract',
    'Dependents',
    'InternetService',
    'PaymentMethod',
    'PaperlessBilling',
    'StreamingMovies',
    'StreamingTV',
    'DeviceProtection',
    'OnlineSecurity',
  'OnlineBackup',
 'TechSupport',
'MultipleLines',
'PhoneService'
]
```

```python
# from imblearn.over_sampling import SMOTE
# from imblearn.pipeline import Pipeline as ImbPipeline
```

```python
x = df[num_cols + cat_cols]
y = df['Churn']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), cat_cols)
    ]
)
```

```python
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    # ('smote', SMOTE(random_state=42)),
    ('model', GradientBoostingClassifier(random_state=42))
])
```

```python
param_grid = {
    'model__n_estimators': [100],
    'model__max_depth': [3],
    'model__learning_rate' : [0.1]
}
```

```python
grid = GridSearchCV(
    estimator=pipeline,
    param_grid=param_grid,
    cv=5,
    scoring='f1',
    n_jobs=-1,
    verbose=2
)
```

```python
grid.fit(x_train, y_train)
```

```python
best_pipeline = grid.best_estimator_
```

```python
y_pred = best_pipeline.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
precision_score(y_test,y_pred,pos_label='yes')
```

```python
precision_score(y_test,y_pred,pos_label='no')
```

```python
recall_score(y_test,y_pred,pos_label='no')
```

```python
recall_score(y_test,y_pred,pos_label='yes')
```

```python
f1_score(y_test,y_pred,pos_label='yes')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
